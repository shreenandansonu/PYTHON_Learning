```python
import pandas as pd
import numpy as np
```

# 1D array 
- `ones`

- `zeros`

- `linespace`

- `arange`
* we can use these functions to create one dimensional array *

```python
# Creating 1D array in python using numpy
a=np.ones(100,dtype=float)
b=np.zeros(100)
c=np.linspace(1,10,100)
d=np.arange(100)
df=pd.DataFrame({"ones":a,"zeros":b,"linespace":c,"arange":d})
df.to_csv("Random.csv",index=False)
```

```python
x=np.array([a,b,c,d])
print(x.shape)
print(x.size)
print(x.d)
x
```
